```python
import pandas as pd
from Bio import SeqIO
import numpy as np
import matplotlib.pyplot as plt
```

```python
indir = '/share/home/zabel/final_clustalo/'
mspa = []
mspb = []
for record in SeqIO.parse(open(indir + 'mpsa_staph_alignment.fasta', 'r'), 'fasta'):
    mspa.append(record)
for record in SeqIO.parse(open(indir + 'mpsb_staph_alignment.fasta', 'r'), 'fasta'):
    mspb.append(record)
```

```python
for a in mspa:
    print(a.description)
```

```python
concatenated_records = []
for a in mspa:
    d_a = a.description
    for b in mspb:
        d_b = b.description
        if d_a==d_b:
            s = a.seq+b.seq
            a.seq = s
            concatenated_records.append(a)
            break
```

```python
output_file = '/share/home/zabel/final_clustalo/concatenated_alignment_staph.fasta'
with open (output_file, 'w') as output_handle:
    SeqIO.write(concatenated_records, output_handle, 'fasta')
```

```python
names = [i.description for i in concatenated_records]
for i in concatenated_records:
    print(i)
    
```
